```
#https://twitter.com/search?q=(%23sarcasm)%20lang%3Ahi&src=typed_query&f=live
import twitterscraper
from twitterscraper import query_tweets
from langdetect import detect

# All tweets matching either Trump or Clinton will be returned. You will get at
# least 10 results within the minimal possible time/number of requests
file = open('output.txt','w')
count = 0
for tweet in query_tweets("(#sarcasm) lang:hi", 100):
    if(tweet.text==''):
      continue
    lang = detect(tweet.text)
    if(lang=="hi" or lang=="mr"):
      continue
    maxchar = max(tweet.text)
    if maxchar > u'\u007f':
      continue
    mystring=tweet.text
    mystring = mystring.replace('\n', ' ').replace('\r', '')
    file.write(tweet.tweet_id+'\n')
    file.write(mystring +'\n') 
    file.write('\n')
    count=count+1
    #print(lang)
    #print(tweet.text)
    #print(tweet.tweet_id)

print(count)
file.close()

#import codecs,json
#twitterscraper --limit 100 --output=tweets.json
#with codecs.open('tweets.json','r','utf-8') as f:
#  tweets=json.load(f,encoding='utf-8')

#print(tweets)
```
