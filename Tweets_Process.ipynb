```


import string
import re
import operator
fp = open("embed.txt", 'r')
stop_words=[]
rare_words=[]
token_count = {}
nc=0
for line in fp:
	line = line.strip()
	line = re.split(r'\s+|[!$%&()*+, -./:;<=>?\^_`{|}~]\s*', line)
	line = [token.strip() for token in line if token != '' and token != ' ' and token != '\n']
	if len(line) > 1:
		for i in range(len(line)):
		  if line[i].lower() in token_count:
			  token_count[line[i].lower()] += 1
		  else:
			  token_count[line[i].lower()] = 1
		nc=nc+1

token_count = dict(sorted(token_count.items(), key=operator.itemgetter(1)))
for key, value in token_count.items():
	if value >= 100:
		stop_words.append(key.lower())
	elif value <5:
		rare_words.append(key.lower())
print(nc)

```

```
id_tokenised_tweet_map = []
#line="Naa bhai.. aisa nhi hai.. mere handle karne se bhi kuchh hona nhi hai.. politics se mera door door tak ka naata nhi hai #wow #omg"
#line = re.split(r'\s+|[,;.-]\s*', line)
#print(line)
fp = open("embed.txt", 'r')
count=0
for line in fp:
	#print(line)
	line = line.strip()
	line = re.split(r'\s+|[!$%&()*+, -./:;<=>?\^_`{|}~]\s*', line)
	#print(line)
	line = [token.strip() for token in line if token != '' and token != ' ' and token != '\n']
	if len(line) > 1:
		tokens = []
		for i in range(len(line)):
			if line[i][0] == '#':
				tokens.append(line[i][1:])
			elif line[i][0] == '@' or line[i][0] in string.punctuation :
			  continue
			elif 'http' in line[i]:
				tokens.append('URL')
			elif line[i].lower() in rare_words :
				continue
			else:
				tokens.append(line[i])
		count=count+1
		#print(tokens)
		id_tokenised_tweet_map.append(tokens) 


```

```
fp = open("large_total_tweets.txt", 'r')
f2=open("large_total_tweets_with_language.txt",'w')
import string
import nltk
nltk.download('wordnet')
#import enchant
#d = enchant.Dict("en_US")
from nltk.corpus import wordnet

for line in fp:
  #print(line)
  line = line.strip()
  line = line.split(' ')
  #line = re.split(r'\s+|[!$%&()*+, -./:;<=>?\^_`{|}~]\s*', line)
  #print(line)
  line = [token.strip() for token in line if token != '' and token != ' ' and token != '\n']
  if len(line) > 1:
    #print(line)
    for i in range(len(line)):
      if line[i] == '' or line[i] == ' ' or line[i] == '\n' :
        continue
      elif line[i][0] == '#':
        if len(line[i])>1:
          f2.write(line[i][1:]+'      '+'rest'+'\n')
      elif line[i][0] == '@' or line[i][0] in string.punctuation :
        f2.write(line[i][0:]+'      '+'rest'+'\n')
      elif 'http' in line[i]:
        f2.write(line[i][0:]+'      '+'rest'+'\n')
      else:
        if not wordnet.synsets(line[i]):
          f2.write(line[i]+'      '+'hi'+'\n')
        else:
          f2.write(line[i]+'      '+'en'+'\n')
    f2.write('\n')
  elif len(line)==1:
    f2.write(line[0]+'\n')
f2.close()
```
